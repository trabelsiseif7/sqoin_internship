```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import requests
from bs4 import BeautifulSoup

```

```python
def scrape_quora_url(url):
    driver = webdriver.Chrome()  # Replace with the appropriate webdriver for your browser
    driver.get(url)

    PAUSE_TIME = 2

    lh = driver.execute_script("return document.body.scrollHeight")

    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(PAUSE_TIME)
        nh = driver.execute_script("return document.body.scrollHeight")
        if nh == lh:
            break
        lh = nh

    questions = driver.find_elements(By.CSS_SELECTOR, 'div.q-box.qu-mb--tiny')
    answers = driver.find_elements(By.CSS_SELECTOR, 'div.q-box.spacing_log_answer_content.puppeteer_test_answer_content')

    question_texts = [question.text for question in questions]
    answer_texts = [answer.text.replace('\n', ' ') for answer in answers]

    driver.quit()

    return question_texts, answer_texts
```

```python
keyword = "blockchain"
num_results = 10
url = f"https://www.google.com/search?q={keyword}+site:quora.com&num={num_results}"

requests_results = requests.get(url)
soup_link = BeautifulSoup(requests_results.content, "html.parser")
links = soup_link.find_all("a")
```

```python
data = []
for link in links:
    link_href = link.get('href')
    if "url?q=" in link_href and not "webcache" in link_href:
        title = link.find_all('h3')
        if len(title) > 0:
            url = link.get('href').split("?q=")[1].split("&sa=U")[0]
            title_text = title[0].getText()
            data.append({
                'URL': url,
                'Title': title_text
            })
```

```python
data
```

```python
df = pd.DataFrame(columns=['URL', 'Question', 'Answer'])
for entry in data:
    question_texts, answer_texts = scrape_quora_url(entry['URL'])
    for question, answer in zip(question_texts, answer_texts):
        df = df.append({"URL": entry['URL'], "Question": question, "Answer": answer}, ignore_index=True)
```
